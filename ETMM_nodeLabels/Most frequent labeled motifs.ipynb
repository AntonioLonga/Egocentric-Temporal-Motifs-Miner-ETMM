```python
import construction as cs
import ETN as etn
import numpy as np
import networkx as nx
import itertools

%load_ext autoreload
%autoreload 2
```

```python
def one_function(path,meta_path,gap,top_k=14):
    
    df = cs.load_data(path)
    graphs = cs.build_graphs(df,gap,with_labels=True,meta_path=meta_path)
    k=3

    meta = cs.load_metadata(meta_path)
    S = etn.count_ETN(graphs,k,meta)
    S = {k: v for k, v in sorted(S.items(), key=lambda item: item[1],reverse=True)}

    etn.draw_ETNs(S,k,meta,n_per_row=7,top_k=top_k)
```

```python
meta_path = "../Datasets/metadata/metadata_LH10.dat"
path = "../Datasets/LH10.dat"
one_function(path,meta_path,299,top_k=14)
```
